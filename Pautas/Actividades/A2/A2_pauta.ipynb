## Ejercicio 1

```python
def bodegas_vacias(k, eventos):
  right_stack = [k - 1]
  left_stack = [0]
  state = [0] * k
  for evento in eventos:
    if evento == "I":
      next = left_stack[-1] + 1
      pos = left_stack.pop()
      state[pos] = 1
      left_stack.append(next)
    elif evento == "D":
      next = right_stack[-1] - 1
      pos = right_stack.pop()
      state[pos] = 1
      right_stack.append(next)
    else:
      pos = int(evento)
      state[pos] = 0
      if pos < left_stack[-1]:
        left_stack.append(pos)
      else:
        right_stack.append(pos)
  return state
```

```python
if __name__ == "__main__":
  k = 10
  eventos = "I0I0IIDD9"
  num_bodegas = bodegas_vacias(k, eventos)
  print(num_bodegas)
```

## Ejercicio 2

```python
def pares_misma_diferencia(lista):
  restado = [elem - i for i, elem in enumerate(lista)]
  cuentas = {}

  for elem in restado:
    if elem not in cuentas:
      cuentas[elem] = 0
    cuentas[elem] += 1
    
  total_pares = 0
  for cuenta in cuentas.values():
    total_pares += int(cuenta*(cuenta-1)/2)

  return total_pares
```

```python
if __name__ == "__main__":
  lista = [4, 5, 6, 7, 8, 9]
  num_pares = pares_misma_diferencia(lista)
  print(num_pares)
```

## Ejercicio 3

```python
def dfs(nodo, grafo, visitados):
  if visitados[nodo]:
    return 0
  visitados[nodo] = True
  respuesta = 1
  for vecino in grafo[nodo]:
    respuesta += dfs(vecino, grafo, visitados)

  return respuesta

def difusion_epifania(grupos):
  n = -1
  for grupo in grupos:
    n = max(n, *grupo)
  n += 1

  grafo = [[] for _ in range(n)]
  visitados = [0 for _ in range(n)]

  for grupo in grupos:
    for i, persona in enumerate(grupo[1:], start=1):
      grafo[persona].append(grupo[i - 1])
      grafo[grupo[i-1]].append(persona)
  print(grafo)
  mejor_conectividad = -1
  mejor_nodo = -1
  for nodo in range(n):
    conectividad = dfs(nodo, grafo, visitados)
    #print(nodo, conectividad)
    if conectividad > mejor_conectividad:
      mejor_conectividad = conectividad
      mejor_nodo = nodo

  return mejor_nodo
```

```python
if __name__ == "__main__":
  grupos = [[1,4,3],[0,1],[0],[5,6],[2]]
  mejor_nodo = difusion_epifania(grupos)
  print(mejor_nodo)
```
